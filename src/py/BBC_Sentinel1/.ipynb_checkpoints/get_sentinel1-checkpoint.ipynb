```python
from sentinelsat import SentinelAPI, read_geojson, geojson_to_wkt
from datetime import date
```

```python
user="adfraser"
password='fuckidunno0)'
#I used my own credentials, you should be able to sign up can use your own

footprint=geojson_to_wkt(read_geojson('/scratch/cloudstor/BBC_Sentinel1/Mawson.geojson')) # coorinates are in this file in a specific format

query_kwargs = {
        'platformname': 'Sentinel-1',
        'producttype': 'GRD',
        'date': ('2018-08-06T00:00:00.000Z','2018-08-15T23:59:00.000Z' ),
        'instrumentshortname': 'SAR-C SAR',
        'sensoroperationalmode': 'EW'
}

api = SentinelAPI(user, password, 'https://scihub.copernicus.eu/dhus')

# this will return how many files matched the search
count = api.count(footprint,**query_kwargs)

# this returns metadata about the files so that can filter them more 
products = api.query(footprint,**query_kwargs)   

products_df = api.to_dataframe(products)

#products_df.set_option('display.max_colwidth', -1)

#print(count)
#print(products_df.title[0:19])
#print(products_df.title[20:39])
#print(products_df.title[40:59])
#print(products_df.title[60:79])
#print(products_df.title[80:83])

for file in products:
    print(products_df.title[file])

```
